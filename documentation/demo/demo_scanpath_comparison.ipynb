```python
import Vision as v
import os
import numpy as np
```

```python
data_1 = 'dataset/nat006.csv'  
data_2 = 'dataset/nat006_ter.csv' 

image_ref = 'dataset/nat006.bmp'
```

```python
sp_1 = v.Scanpath(data_1, 
                sampling_frequency = 500,  
                segmentation_method = 'I_HMM',
                distance_type = 'euclidean',                         
                display_segmentation = False,
                size_plan_x = 921,
                size_plan_y = 630,  
                ref_image=image_ref,
                display_scanpath=True, 
                verbose=False)
```

```python
sp_2 = v.Scanpath(data_2, 
                  sampling_frequency = 500,  
                  segmentation_method = 'I_HMM',
                  distance_type = 'euclidean',                         
                  display_segmentation = False,
                  size_plan_x = 921,
                  size_plan_y = 630,  
                  ref_image=image_ref,
                  display_scanpath=True, 
                  verbose=False)
```

```python
v.DTW_distance([sp_1, sp_2], 
              verbose=False)
```

```python
v.subsmatch_similarity([sp_1, sp_2], 
                      verbose=False)
```

```python
v.multimatch_alignment([sp_1, sp_2], 
                      verbose=False)
```
