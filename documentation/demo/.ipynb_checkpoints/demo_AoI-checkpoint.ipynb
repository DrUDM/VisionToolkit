```python
import vision_toolkit as v
import os
import numpy as np
```

```python
data = 'demo_data/nat006.csv' 
image_ref = 'demo_data/nat006.bmp'
```

```python
bs = v.BinarySegmentation(data, 
                          sampling_frequency = 500,  
                          segmentation_method = 'I_HMM',
                          distance_type = 'euclidean',                        
                          display_segmentation = False,
                          verbose=False,
                          size_plan_x = 921,
                          size_plan_y = 630,  
                          )
```

```python
aoi_s = v.AoISequence(bs, 
                      AoI_identification_method='I_KM', 
                      AoI_IKM_cluster_number='search', 
                      verbose=True)
```

```python
aoi_s = v.AoISequence(bs, 
                      ref_image=image_ref,
                      AoI_identification_method='I_AP',  
                      verbose=False)
```

```python
v.AoI_transition_matrix(aoi_s)
```

```python
v.AoI_transition_entropy(aoi_s)
```

```python
v.AoI_NGram(aoi_s)
```

```python
v.AoI_lempel_ziv(aoi_s)
```
