```python
import Vision as v
import os
import numpy as np
```

```python
data = 'dataset/nat006.csv' 
image_ref = 'dataset/nat006.bmp'
```

```python
bs = v.BinarySegmentation(data, 
                          sampling_frequency = 500,  
                          segmentation_method = 'I_HMM',
                          distance_type = 'euclidean',                         
                          display_segmentation = False,
                          size_plan_x = 921,
                          size_plan_y = 630,  
                          verbose=False)
```

```python
sp = v.Scanpath(bs, 
                ref_image=image_ref,
                display_scanpath=True, 
                verbose=True)
```

```python
sp = v.Scanpath(data, 
                sampling_frequency = 500,  
                segmentation_method = 'I_HMM',
                distance_type = 'euclidean',                         
                display_segmentation = False,
                size_plan_x = 921,
                size_plan_y = 630,  
                ref_image=image_ref,
                display_scanpath=True, 
                verbose=False)
```

```python
v.scanpath_convex_hull(sp,
                       display_results=True)
```

```python
v.scanpath_voronoi_cells(sp, 
                         display_results=True, 
                         get_raw=False)
```

```python
v.scanpath_HFD(sp,
               display_results=True)
```
